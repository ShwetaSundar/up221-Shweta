```python
# importing geopandas to plot map
import geopandas as gpd
```

```python
# importing a LA county census tract geojson file from the class data
tracts = gpd.read_file('Census_Tracts_2020.geojson')
tracts.head()
```

```python
# plotting geojson file
tracts.plot(figsize = (12,10))
```

```python
# what type of data in this dataset 
tracts.info(verbose=True, show_counts=True)
```

```python
# extracting only FIPS and geometry data from the data
tracts = tracts [['CT20', 'geometry']]
tracts.head()
                  
```

```python
#creating a FIPS column to be able to import other data into this dataset
tracts['FIPS'] = '06'+'037'+tracts['CT20']
tracts.head()
```

```python
tracts= tracts[['FIPS','geometry']]
tracts.head()
```

```python
#merging the geojson dataset with the dataset with housing units
tracts_housing = tracts.merge(df2, on='FIPS')
```

```python
tracts_housing.head()
```

```python
# plotting column with number of housing units per census tract
tracts_housing.plot(figsize = (12,15),
                     column = 'No_Housing_Units',
                     legend = True, 
                     scheme ='natural_breaks')
```

```python
# filtering top 10 censts tracts with highest number of houses
tracts_housing_high = tracts_housing.sort_values(by= 'No_Housing_Units', ascending = False)
```

```python
tracts_housing_high = tracts_housing_high.iloc[0:10]
tracts_housing_high.shape
```

```python
# plotting top 10 censts tracts with highest number of houses
tracts_housing_high.plot(figsize = (12,10),
                     column = 'No_Housing_Units',
                     legend = True, 
                     scheme ='natural_breaks')
```

```python
import folium
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts_housing, # geo data
                  data=tracts_housing, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS', 'No_Housing_Units'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Distribution of housing units (2020)').add_to(m)    # name on the legend color bar
```
